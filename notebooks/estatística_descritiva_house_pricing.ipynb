### **Relembrando algumas medidas estatísticas importantes:**

As medidas de tendência central são:

- **Media**

Dados os $n$ números ${\{x_{1}, x_{2}, \ldots, x_{n}\}}$, a média aritmética é definida como:

$$
\begin{equation}
    \bar {x} = {\frac{1}{n}} \sum_{i = 1}^{n}x_{i} ={\frac{x_{1} + x_{2} + \cdots + x_{n}}{n}}
\end{equation}
$$

- **Mediana**

Sejam ${x_{1}, x_{2}, x_{3}, \ldots, x_{n}}$ os dados de uma amostra ordenada em ordem crescente e designando a mediana como $M_{e}$, distinguimos dois casos:

a) Se $n$ for ímpar, a mediana é o valor que ocupa a posição $\frac{(n + 1)}{2}$ uma vez ordenados os dados, em ordem crescente ou decrescente, pois este é o valor central. Ou seja: $M_{e}=x_{{(n+1)/2}}$.

b) Se $n$ for par, a mediana é a média aritmética dos dois valores centrais. Quando $n$ é par, os dois dados no centro da amostra ocupam as posições $\frac{n}{2}$ e $ \frac{n}{2} + 1$. Ou seja: $M_{e}=(x_{{{n / 2}}} + x_{{{{(n / 2)}}+1}})/2$.

-  **Moda**

A moda é o valor dos dados com mair freqüência em um conjunto.



*É importante ressaltar que com o bom uso da estatística evita quaisquer tipo de generalização ou inferência além do que está disponível. Isso significa que as estatísticas descritivas são apenas a representação dos dados (amostra) disponíveis e não se baseiam em nenhuma teoria de probabilidade.*

### **Quartis e boxplot**

Quartis são valores que dividem uma amostra de dados em quatro partes iguais. Com eles você pode rapidamente avaliar a dispersão e a tendência central de um conjunto de dados, que são etapas importantes na compreensão dos seus dados.

**3º quartil (Q3):**	75% dos dados são menores que ou iguais a este valor.

**Amplitude interquartílica:**	A distância entre o 1o. e o 3o. quartis (Q3-Q1); assim, ele ultrapassa o meio de 50% dos dados.

# Correlação e Gráficos de Dispersão
A correlação é uma medida estatística que descreve a força e a direção de um relacionamento linear entre duas variáveis. Quando queremos entender como duas variáveis se relacionam entre si, podemos usar gráficos de dispersão e calcular coeficientes de correlação.

## Gráficos de Dispersão
Um gráfico de dispersão (scatter plot) é uma representação visual de duas variáveis. Cada ponto no gráfico representa uma observação do conjunto de dados. A posição do ponto é determinada pelos valores das duas variáveis. Por exemplo, se quisermos analisar a relação entre o preço de venda de imóveis (SalePrice) e a qualidade geral (OverallQual), cada ponto no gráfico representaria um imóvel, com o eixo x mostrando o preço e o eixo y mostrando a qualidade.

## Coeficientes de Correlação
Os coeficientes de correlação quantificam a força e a direção da relação entre duas variáveis. Os valores variam de -1 a 1:

- 1: Correlação positiva perfeita (à medida que uma variável aumenta, a outra também aumenta).
- 0: Nenhuma correlação linear (não há relacionamento linear entre as variáveis).
- -1: Correlação negativa perfeita (à medida que uma variável aumenta, a outra diminui).

No contexto de análise de dados de imóveis, a correlação pode nos ajudar a entender, por exemplo, se há uma relação entre o preço de venda de um imóvel e sua qualidade, ou o ano de construção, ou o número de quartos.

## **Dispersão**


### **Relembrando conceitos importantes**

1. Distribuição Normal

Uma distribuição normal, também conhecida como curva gaussiana, é uma distribuição simétrica em forma de sino, onde a maioria dos dados se concentra ao redor da média, e a probabilidade de valores extremos diminui gradualmente à medida que nos afastamos da média. Muitos testes estatísticos assumem que os dados seguem uma distribuição normal, pois isso permite a aplicação de métodos paramétricos.

2. Desvio Padrão

O desvio padrão é uma medida de dispersão que indica o quanto os valores de um conjunto de dados variam em relação à média. Um desvio padrão alto indica que os dados estão mais espalhados, enquanto um desvio padrão baixo indica que os dados estão mais concentrados em torno da média.

3. Curtose

A curtose mede a "espessura" das caudas de uma distribuição em comparação com uma distribuição normal. Uma curtose positiva indica caudas mais pesadas (mais valores extremos), enquanto uma curtose negativa indica caudas mais leves (menos valores extremos).

4. Assimetria (Skew)

A assimetria (ou skew) mede a simetria de uma distribuição. Uma distribuição simétrica tem um skew de zero. Uma distribuição com cauda mais longa à direita (cauda positiva) tem skew positivo, enquanto uma distribuição com cauda mais longa à esquerda (cauda negativa) tem skew negativo.

### Teste t
O teste t é um teste estatístico utilizado para comparar as médias de dois grupos e determinar se elas são significativamente diferentes entre si. Existem diferentes tipos de testes t, como:

*  Teste t para amostras independentes: Compara as médias de dois grupos independentes (por exemplo, casas caras e casas baratas).
*  Teste t pareado: Compara as médias de dois grupos relacionados (por exemplo, antes e depois de um tratamento).
*  Teste t para uma amostra: Compara a média de um grupo com um valor conhecido.

Para usar o teste t comparando a média de OverallQual para as casas caras e baratas, precisamos, antes de mais nada, descobrir se as variâncias são iguais, pois temos testes t diferentes para cada caso. Se as variâncias são iguais, usamos o teste t para variâncias iguais. Caso contrário, usamos o teste t para variâncias desiguais.

Para determinar se as variâncias de dois conjuntos de dados são iguais, podemos usar o teste F, que é adequado quando temos mais de 20 observações em cada conjunto. O teste F é um teste estatístico que compara as variâncias de duas amostras para verificar se elas são significativamente diferentes. A seguir, estão os passos e as hipóteses envolvidas no teste F:

1. **Formulação das Hipóteses:**
   - **Hipótese Nula (H0):** As variâncias dos dois conjuntos de dados são iguais.
   - **Hipótese Alternativa (Ha):** As variâncias dos dois conjuntos de dados são diferentes.

2. **Cálculo da Estatística F:**
   - A estatística F é calculada como a razão entre as variâncias das duas amostras. Se \(s_1^2\) e \(s_2^2\) representam as variâncias das duas amostras, então a estatística F é dada por:
     \[
     F = s_1^2/s_2^2
     \]

3. **Determinação do Valor Crítico e Comparação:**
   - O valor crítico de F é determinado a partir da distribuição F com os graus de liberdade apropriados para as duas amostras. Este valor é comparado com a estatística F calculada para decidir se rejeitamos ou não a hipótese nula.

4. **Conclusão:**
   - Se a estatística F calculada for maior que o valor crítico, rejeitamos a hipótese nula (H0) e concluímos que as variâncias são significativamente diferentes.
   - Se a estatística F calculada for menor ou igual ao valor crítico, não rejeitamos a hipótese nula (H0), indicando que não há evidências suficientes para afirmar que as variâncias são diferentes.

Portanto, ao aplicar o teste F com mais de 20 observações em cada grupo, estamos verificando se há uma diferença significativa entre as variâncias dos dois conjuntos de dados.

````python
df.groupby("categoria_preco")["OverallQual"].var()
````

````python
import scipy.stats

#qualidade geral apenas das caras
caras = df[df.categoria_preco == "cara"].OverallQual

#qualidade geral apenas das baratas
baratas = df[df.categoria_preco == "barata"].OverallQual

df_cara = len(caras) - 1 #grau de liberdade da primeira amostra

df_barata = len(baratas) - 1 #grau de liberdade da segunda amostra

F = baratas.var() / caras.var() #A maior variância tem que ir no numerador!

alpha = 0.05

#p-valor para o teste F
1 - scipy.stats.f.cdf(F, df_barata, df_cara)
````

p-valor é maior do que alpha (0.05), portanto podemos considerar as variancias iguais!

A ANOVA (Análise de Variância) é uma técnica estatística utilizada para verificar se as médias de três ou mais grupos são significativamente diferentes entre si. Para que a ANOVA seja aplicada corretamente, devem ser atendidas as seguintes condições:

1. **Amostras Aleatórias:**
   - As amostras de cada grupo devem ser selecionadas aleatoriamente da população, garantindo que cada membro tenha uma chance igual de ser incluído.

2. **Grupos Independentes:**
   - Os grupos devem ser independentes uns dos outros, ou seja, a seleção de membros de um grupo não deve influenciar a seleção de membros de outro grupo.

3. **Variável Dependente Contínua:**
   - A variável que estamos analisando deve ser contínua, o que significa que pode assumir qualquer valor dentro de um intervalo.

4. **Tamanho das Amostras:**
   - Cada grupo deve ter pelo menos 15 observações (n ≥ 15), garantindo que as estimativas sejam suficientemente precisas.

5. **Homogeneidade das Variâncias:**
   - As variâncias populacionais dos diferentes grupos devem ser aproximadamente iguais, uma condição conhecida como homocedasticidade. Isso assegura que a comparação entre as médias seja válida.

Atendendo a esses critérios, a ANOVA pode ser usada para testar a hipótese nula de que todas as médias dos grupos são iguais contra a hipótese alternativa de que pelo menos uma média é diferente das outras.



### Teste de Bartlett

O teste de Bartlett é utilizado para verificar se as variâncias de múltiplas amostras são iguais. A seguir, detalham-se as hipóteses e condições para o uso desse teste:

1. **Hipóteses do Teste:**
   - **Hipótese Nula (H0):** As variâncias são iguais para todas as amostras.
   - **Hipótese Alternativa (Ha):** As variâncias não são iguais para pelo menos um par de amostras.

2. **Condições para Uso:**
   - **Distribuição Normal dos Dados:** O teste de Bartlett pressupõe que os dados sejam normalmente distribuídos em cada amostra.
   - **Tamanho da Amostra:** Pode ser usado de forma mais robusta quando cada grupo tem mais de 20 observações.

Quando essas condições são atendidas, o teste de Bartlett é uma ferramenta eficaz para verificar a homogeneidade das variâncias, o que é um requisito essencial para muitas análises estatísticas, como a ANOVA.

* A hipótese nula para o teste é que as variâncias são iguais para todas as amostras. Em termos estatísticos, isso é:
H0: σ12 = σ22 = … = σk2.

* A hipótese alternativa é que as variâncias não são iguais para um par ou mais:
Ha: σ12 ≠ σ22 ≠… ≠ σk2.
* Pode ser usada se tiver dados normalmente distribuídos ou +20 amostras

O teste Kruskal-Wallis é um teste não paramétrico que pode ser usado para determinar se existem diferenças estatisticamente significativas entre dois ou mais grupos de uma variável independente em um variável dependente contínua ou ordinal. É considerada a alternativa não paramétrica à ANOVA one-way e uma extensão do teste U de Mann-Whitney para permitir a comparação de mais de dois grupos independentes.

````python
stats.kruskal(list_RL, list_C, list_FB, list_RH, list_RM)
````
